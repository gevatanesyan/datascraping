```python
import scrapy
from scrapy.crawler import CrawlerProcess
from scrapy.http import TextResponse
import requests
import numpy as np
import pandas as pd
from scrapy.http import TextResponse
import re
from selenium import webdriver
```

```python
class Menu(scrapy.Spider):
    name = 'menu'
    start_urls = ['https://www.menu.am/?status=all&sort=default']
    
    custom_settings = {
      "FEED_FORMAT":"json",
      "FEED_URI":"menu.json",
      "ROBOTSTXT_OBEY":True,
    }
    
    def parse(self,response):
        x =response.css('div.restaurants_grid')
        k = x.css('div.new_favorites_and_rates_block')

        Name = x.css('div.item::attr(data-marker_title)').extract()
        Time = x.css('span.new_list_time_block_inner::text').extract()
        Url = [response.urljoin(k) for k in (x.css('a.title::attr(href)').extract())]
        Category = x.css('span.restType::text').extract()
        Rate = [i.css('div.new_rates_block::text').extract_first(default = None) for i in k]
        
        result = []
        for Name, Time, Url, Category, Rate in zip(Name, Time, Url, Category, Rate):
            result.append({
                'Name': Name,
                'Time': Time,
                'Url': Url,
                'Category': Category,
                'Rate': Rate
            })
        return result
        #return {'Name':Name,'Time':Time,'Url':Url,'Category':Category,'Rate':Rate}
            
        
        
```

```python
process = CrawlerProcess()
process.crawl(Menu)
process.start()
  
```
