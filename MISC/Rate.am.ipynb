```python
def ratescraper():
    result = {}
    PATH = r"/Users/gevorgatanesyan/Downloads/chromedriver"
    URL = "http://rate.am/am/armenian-dram-exchange-rates/central-bank-armenia"
    currency = [' 1 USD (Ամերիկյան դոլար)', ' 1 EUR (ԵՎՐՈ)', ' 1 RUR (Ռուսաստանյան ռուբլի)',' 1 GBP (Մեծ Բրիտանիայի ֆունտ ստերլինգ)']
    browser = webdriver.Chrome(PATH)
    browser.get(URL)
    for item in currency:
        time.sleep(2)
        data = []
        currid = browser.find_element_by_name('ctl00$Content$dlCurrency')
        time.sleep(2)
        currid.send_keys(item)
        time.sleep(2)
        currid = browser.find_element_by_name('ctl00$Content$dlCurrency')
        time.sleep(2)
        currid.send_keys(item)
        for i in range(2020,2017,-1):
            curryear = browser.find_element_by_name('ctl00$Content$dlYear')
            time.sleep(2)
            curryear.send_keys(i)
            time.sleep(2)
            curryear = browser.find_element_by_name('ctl00$Content$dlYear')
            time.sleep(2)
            curryear.send_keys(i)
            time.sleep(2)
            page = browser.page_source
            data.append(pd.read_html(page)[4][:31])
            result[item] = pd.concat(data)
    browser.close()
    return result 
```

```python
ratedata = ratescraper()
```

### Task 1

```python
annual_vars = []
for k, _ in ratedata.items():
    ratedata[k] = ratedata[k].replace("Կիրակի", np.nan).replace("X", np.nan).apply(pd.to_numeric, errors='coerce')
    df = ratedata[k]
    var = {
        '2019': df.iloc[31:62, 1:].melt().std()['value'] ** 2,
        '2018': df.iloc[62:, 1:].melt().std()['value'] ** 2
    }
    annual_vars.append({'currency': k.strip(), 'variances': var})
```
